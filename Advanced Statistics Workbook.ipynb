```python
#getting quartiles#
for i in np.linspace(0,5,100000):
    if (owl_cdf(i)>=0.24 and owl_cdf(i)<=0.26):
        q1=i
    elif (owl_cdf(i)>=0.49 and owl_cdf(i)<=0.51):
        q2=i
    elif (owl_cdf(i)>=0.74 and owl_cdf(i)<=0.76):
        q3=i
```
